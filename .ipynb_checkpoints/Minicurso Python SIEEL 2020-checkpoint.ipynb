```python
def fibonacci(n):
    x = [0,1]
    for i in range(1,n-1):
        x.append(x[i-1]+x[i])
    print(x)
    for i in range(len(x)):
        if i==len(x)-1:
            print(i)
        else:
            print(i, end=', ')
        
```

## Trabalhando com dados da Apple, Microsoft, IBM e Google

```python
def get_data(tickers_list, start_date, end_date):
    def data(ticker):
        return (pdr.get_data_yahoo(ticker, start=start_date, end=end_date))
    datas = map(data, tickers_list)
    return(pd.concat(datas, keys=tickers_list, names=['Ticker', 'Date']))
```

```python
tickers_list = ['AAPL', 'MSFT', 'IBM', 'GOOG']
all_data = get_data(tickers_list, datetime.datetime(2006, 10, 1), datetime.datetime(2012, 1, 1))
all_data
```

```python
# Altera o DataFrame de forma que o índice seja 'Date', as colunas 'Ticker' e os valores 'Adj Close'
daily_close_px = all_data[['Adj Close']].reset_index().pivot('Date', 'Ticker', 'Adj Close')

# Calcula a mudança percentual diária para 'daily_close_px'
daily_pct_change = daily_close_px.pct_change()

# Plota os histogramas
daily_pct_change.hist(bins=50, sharex=True, figsize=(12, 8))
plt.show()
```

```python
# Plota uma matriz de dispersão de 'daily_pct_change'
pd.plotting.scatter_matrix(daily_pct_change, diagonal='kde', alpha=0.1,figsize=(12, 12))
plt.show()
```

## Janela Móvel

- Funciona como um filtro passa-baixa filtrando ocilações na métrica medida e evidenciando tendencias

## Volatilidade

- Mede a mudança da variancia do retorno sobre um periodo de tempo
- Utilizado para comparar risco
- Quanto maior a volatilidade maior o risco

```python
# Define o tamanho da janela da avaliação
min_periods = 75 

# Calcula a volatilidade
vol = daily_pct_change.rolling(min_periods).std() * np.sqrt(min_periods) 

vol.plot(figsize=(12, 8))
plt.show()
```

## Regressão de Mínimos Quadrados

- Busca modelar a relação entre variaveis
- DF Residuals = Graus de liberdade dos residuos
- R-squareds = Indica o quão bem a regressão aproxima os dados
- F-statistic = Mede quão significativo é a regressão

```python
# Isolando o preço de fechamento ajustado
all_adj_close = all_data[['Adj Close']]

# Calcula os ganhos
all_returns = np.log(all_adj_close / all_adj_close.shift(1))

# Isola os valores da AAPL
aapl_returns = all_returns.iloc[all_returns.index.get_level_values('Ticker') == 'AAPL']
aapl_returns.index = aapl_returns.index.droplevel('Ticker')

# Isola os valores da MSFT
msft_returns = all_returns.iloc[all_returns.index.get_level_values('Ticker') == 'MSFT']
msft_returns.index = msft_returns.index.droplevel('Ticker')

# Constroi um novo DataFrame com os ganhos da AAPL e MSFT
return_data = pd.concat([aapl_returns, msft_returns], axis=1)[1:]
return_data.columns = ['AAPL', 'MSFT']

# Adiciona uma constante (eixo x para o modelo)
X = sm.add_constant(return_data['AAPL'])

# Constroi o modelo (procura modelar a relação entre ganhos da AAPL e da MSFT)
model = sm.OLS(return_data['MSFT'],X).fit()

print(model.summary())
```

```python
plt.figure(figsize=(12, 8))

# Plota os ganhos da AAPÇ e MSFT
plt.plot(return_data['AAPL'], return_data['MSFT'], 'r.')

# Adiciona um eixo ao plot
ax = plt.axis()

# Cria um vetor x para mostrar o modelo
x = np.linspace(ax[0], ax[1] + 0.01)

# Plota a linha de regressão
plt.plot(x, model.params[0] + model.params[1] * x, 'b', lw=2)

plt.grid(True) # Adiciona o grid
plt.axis('tight') # Reduz os espaços em branco
plt.xlabel('Apple Returns') # Nomeia o eixo x
plt.ylabel('Microsoft returns') # Nomeia o eixo y
plt.show()
```

```python
# Plota a correlação móvel entre ganhos da AAPL e da MSFT
return_data['MSFT'].rolling(window=252).corr(return_data['AAPL']).plot(figsize=(12, 8))
plt.show()
```

## Estratégia de Negociação

- Estatégia simples
    - Duas médias moveis
    - Quando a curta ultrapassar a longa = compra
    - Quando a longa ultrapassar a curta = vende

```python
# Define os tamanhos das janelas
short_window = 40
long_window = 100

# Cria o DataFrame 'signals' com o index da AAPL e com a coluna 'signal' inicializada com 0.0
signals = pd.DataFrame(index=aapl.index)
signals['signal'] = 0.0

# Cria a janela curta
signals['short_mavg'] = aapl['Close'].rolling(window=short_window, min_periods=1, center=False).mean()

# Cria a janela longa
signals['long_mavg'] = aapl['Close'].rolling(window=long_window, min_periods=1, center=False).mean()

# Preenche a coluna 'signal' com a indicação de quando 'short_mavg'>'long_mavg'
signals['signal'][short_window:] = np.where(signals['short_mavg'][short_window:]>signals['long_mavg'][short_window:], 1.0, 0.0)   

# Faz a diferença do sinal atual com o anterior para criar a ordem
#  1 - Compra
#  0 - Não faz nada
# -1 - Vende
signals['positions'] = signals['signal'].diff()
print(signals)
```

```python
fig = plt.figure(figsize=(12, 8))
ax1 = fig.add_subplot(111,  ylabel='Price in $')

# Plota o preço de fechamento em vermelho
aapl['Close'].plot(ax=ax1, color='r', lw=2.)

# Plota os sinais das janelas
signals[['short_mavg', 'long_mavg']].plot(ax=ax1, lw=2.)

# Plota o sinal de compra (triangulo magenta)
ax1.plot(signals.loc[signals.positions == 1.0].index, 
         signals.short_mavg[signals.positions == 1.0],
         '^', markersize=10, color='m')
         
# Plota o sinal de venda (triangulo inverido preto)
ax1.plot(signals.loc[signals.positions == -1.0].index, 
         signals.short_mavg[signals.positions == -1.0],
         'v', markersize=10, color='k')
plt.show()
```

```python
# Define o capital inicial
initial_capital= float(100000.0)

# Cria o DataFrame 'positions'
positions = pd.DataFrame(index=signals.index).fillna(0.0)

# Compra 100 ações
positions['AAPL'] = 100*signals['signal']   
  
# Inicializa o portifolio com o valor possuído
portfolio = positions.multiply(aapl['Adj Close'], axis=0)

# Armazena a diferença em ações possuídas
pos_diff = positions.diff()

# Adiciona 'holdings' ao portifolio
portfolio['holdings'] = (positions.multiply(aapl['Adj Close'], axis=0)).sum(axis=1)

# Adiciona 'cash' ao portifolio
portfolio['cash'] = initial_capital - (pos_diff.multiply(aapl['Adj Close'], axis=0)).sum(axis=1).cumsum()   

# Adiciona 'total' ao portifolio
portfolio['total'] = portfolio['cash'] + portfolio['holdings']

# Adiciona 'returns' ao portifolio
portfolio['returns'] = portfolio['total'].pct_change()
print(portfolio.head())
```

```python
fig = plt.figure(figsize=(12, 8))
ax1 = fig.add_subplot(111, ylabel='Portfolio value in $')

# Plota o portifolio total em dolares
portfolio['total'].plot(ax=ax1, lw=2.)

# Plota o sinal de compra (triangulo magenta)
ax1.plot(portfolio.loc[signals.positions == 1.0].index, 
         portfolio.total[signals.positions == 1.0],
         '^', markersize=10, color='m')

# Plota o sinal de compra (triangulo invertido preto)
ax1.plot(portfolio.loc[signals.positions == -1.0].index, 
         portfolio.total[signals.positions == -1.0],
         'v', markersize=10, color='k')
plt.show()
```

## Sharpe ratio

- Mostra se o retorno obtido é resultado de boas decisões de investimento ou de tomar grandes riscos
- Usualmente para investidores um valor maior que:
    - 1 é aceitável
    - 2 é muito bom
    - 3 é excelente

```python
# Isola os retornos da estrategia
returns = portfolio['returns']

# Calcula o Sharpe ratio anual - 252 dias úteis no ano
sharpe_ratio = np.sqrt(252) * (returns.mean() / returns.std())
print('Sharpe ratio: {}'.format(sharpe_ratio))
```

## Maximum Drawdown

- Mede a maior queda do pico ao fundo no valor de um portfólio
- Indica o risco do portifolio dada a estrategia

```python
# Define uma janela de tamanho 252 - 252 dias úteis no ano
window = 252

# Calcula o maximum drawdown na janela para cada dia
rolling_max = aapl['Adj Close'].rolling(window, min_periods=1).max()
daily_drawdown = aapl['Adj Close']/rolling_max - 1.0

# Calcula o mínimo drawdown diario
max_daily_drawdown = daily_drawdown.rolling(window, min_periods=1).min()

# Plota os resultados
fig = plt.figure(figsize=(12, 8))
daily_drawdown.plot()
max_daily_drawdown.plot()
plt.show()
```

## Compound Annual Growth Rate (CAGR)

- Taxa constante de retorno por um periodo de tempo
- Mostra o que você realmente obteve ao final do sei periodo de investimento
- $EV$ - valor ao final do investimento
- $BV$ - valor inicial do investimento
- $n$ - tamanho do periodo
- $CAGR = (EV/BV)^{1/n}-1$

```python
# Pega o número de dias em AAPL
days = (aapl.index[-1] - aapl.index[0]).days

# Calcula o CAGR 
cagr = ((((aapl['Adj Close'][-1]) / aapl['Adj Close'][1])) ** (365.0/days)) - 1
print('CAGR: {}'.format(cagr))
```
